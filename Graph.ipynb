```python
new_df_set = pd.concat([pd.Series(row['sender'], row['mids'].split(' '))              
                    for _, row in df_set.iterrows()]).reset_index()
new_df_info = pd.concat([pd.Series(row['mid'], row['recipients'].split(' '))              
                    for _, row in df_info.iterrows()]).reset_index()
```

```python
new_df_set
```

```python
new_df_set.columns = ['mid', 'sender']
new_df_info.columns = ['recipient', 'mid']
new_df_info['mid'] = new_df_info['mid'].astype(str)
```

```python
merge_info = new_df_set.merge(new_df_info, how='inner', left_on='mid', right_on='mid')
```

```python
merge_info2 = merge_info.groupby(['sender', 'recipient'], as_index=False).count()
```

```python
G=nx.from_pandas_dataframe(merge_info2, 'sender', 'recipient', ['mid'])
```

```python
from matplotlib import pylab
import networkx as nx

def save_graph(graph,file_name):
#initialze Figure
    plt.figure(num=None, figsize=(8.27,11.69), dpi=80)
    plt.axis('off')
    fig = plt.figure(1)
    pos = nx.spring_layout(graph)
    nx.draw_networkx_nodes(graph,pos)
    nx.draw_networkx_edges(graph,pos)
    nx.draw_networkx_labels(graph,pos)

    cut = 1.00
    xmax = cut * max(xx for xx, yy in pos.values())
    ymax = cut * max(yy for xx, yy in pos.values())
    plt.xlim(0, xmax)
    plt.ylim(0, ymax)

    plt.savefig(file_name,bbox_inches="tight")
    pylab.close()
    del fig

#Assuming that the graph g has nodes and edges entered
save_graph(G,"my_graph.pdf")

#it can also be saved in .svg, .png. or .ps formats
```
